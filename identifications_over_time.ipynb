```python
import pandas as pd
import altair as alt 
import re
```

```python
file = r'.\Working_Datafiles\All_N_Glycopeptides.csv'
df = pd.read_csv(file)
```

```python
grad_time = [0, 18.3, 100, 101, 110, 110.5, 120, 121, 136]
grad_b = [3, 3, 35, 75, 75, 95, 95, 3, 3]

gradient = pd.DataFrame({'time':grad_time, 'buffer_b':grad_b})
gradient['time'] -= 18.3

line = alt.Chart(gradient).mark_line(color='black', strokeDash=[5, 5]).encode(
    x=alt.X('time:Q', title='Time (min)'),
    y=alt.Y('buffer_b:Q', title='%B')
)

peptides = df[(df.concentration==1)].drop_duplicates(['pep_mods', 'temperature'])

bars = alt.Chart(peptides).mark_bar().encode(
    x=alt.X('rt_min:Q', title='', bin=alt.Bin(extent=[0, 120], step=10)),
    y=alt.Y('count()')
)

chart = alt.HConcatChart()
for temp in peptides.temperature.unique():
    b = bars.transform_filter(alt.datum.temperature==temp)
    chart |= (b+line).properties(title=temp)

chart
```
